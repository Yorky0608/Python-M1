```python
from random import randint


secret = randint(1,10)
guess = randint(1,10)

if guess < secret:
    print("Too low!")
elif guess > secret:
    print("Too high!")
else:
    print("Just right!")
```

```python
small = False
green = True

fruits = {"cherry": {"small": True, "green": False}, "pea": {"small": True, "green": True}, "watermelon": {"small": False, "green": True}, "pumpkin": {"small": False, "green": False}}

for fruit, attributes in fruits.items():
    if attributes.get("small") == small and attributes.get("green") == green:
        print(fruit)

```

```python
guess_me = 7
number = 1

while number:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    number += 1
```
